
Now that we have an idea of what is going on in the dataset, it is time to look at the columns with null values and decide what to do with them. My goal is to keep as much information as possible.
    
    1. Smoking Status
    2. BMI

```python
full.smoking_status.head(10)
```

_______________________________________________________________________________________________________________________________
I decided to map the NaN values to never smoked. On forms that people filled out there was probably an N/A or they just did not answer because they do not smoke.
_______________________________________________________________________________________________________________________________

```python
#Using fillna method to convert nan values to 'never smoked' so we can one_hot encode in preprocessing step
full[['smoking_status']] = full[['smoking_status']].fillna('never smoked')
```

______________________________________________________________________________________________________________________________
In the process of trying to decide what to do with the nulls, I saw some outliers. As a health professional, a 97.6 bmi is improbable. BMI usually tops out at ~45-55. A BMI of 40 is already extreme obesity so there is really no need to use the metric past that, as it won't give you any more detail. Waist to Hip ratio would be a better metric to use in such cases. Either way lets use a boxplot to look at all outliers.
______________________________________________________________________________________________________________________________

```python
sns.boxplot(x=full["bmi"]);
plt.title("Boxplot of BMI");
```

```python
#Checking for how many values people have a bmi above 55. These could be typos or human error. BMI charts using top out at 55
print(np.where(full['bmi']>55))
```

```python
print(np.where(full['bmi']>70))
```

_______________________________________________________________________________________________________________________________
Creating a loop to clean up the bmi column. Loop will go through values in the bmi column, if bmi is over 70, it will multiple that number by .1 to get single digit bmi. I will leave the other bmi with high numbers alone since there are over 100. When it is time to model, I will use the original values and the engineered values to look for differences.
_______________________________________________________________________________________________________________________________

```python
for i, bmi in enumerate(full['bmi']):
    if bmi >= 70:
        full['bmi'][i] = bmi * 0.1
```

```python
full.head(10)
```

```python
#Confirming changes took place
print(np.where(full['bmi']>70))
```

```python
full.bmi.isnull().sum()
```

```python
#Looking at the rest of the missing values
missing_bmi = np.where(full['bmi'].isna())[0]
full.iloc[missing_bmi].head(20)
```

```python
full.bmi.mean()
```

_______________________________________________________________________________________________________________________________
Will use the mean of the bmi column to fill in the NaN values. I would rather not ever dropna unless there are only a few rows with NaN values.
_______________________________________________________________________________________________________________________________

```python
full[['bmi']] = full[['bmi']].fillna(full.bmi.mean())
```
